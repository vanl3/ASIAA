```python
phi_dat = np.loadtxt(f'{path_tapwind_jup_e3}domain_x.dat')
rad_dat  = np.loadtxt(f'{path_tapwind_jup_e3}domain_y.dat')[3:-3]

phi = 2.*np.pi*np.linspace(0,1,len(phi_dat)-1) - np.pi/2.
rad   = 0.5*(rad_dat[:-1] + rad_dat[1:])     #take out 

print(rad_dat.shape)
print(rad.shape)
print(rad_dat[0])
print(rad[0])
print((rad_dat[:-1] + rad_dat[1:]).shape)

nx = len(phi)
ny = len(rad)

P, R = np.meshgrid(phi, rad)
X = R*np.cos(P)
Y = R*np.sin(P)

diff_rad = np.diff(rad_dat)
diff_phi = np.diff(phi_dat)

rad2di = np.tile(rad,(ny,1))
rad2di = np.swapaxes(rad2di,0,1)

rad2d = np.tile(diff_rad,(ny,1))
rad2d = np.swapaxes(rad2d,0,1)
phi2d = np.tile(diff_phi,(nx,1))
phi2d = np.swapaxes(phi2d,0,1)

omega = np.sqrt(1/rad2di**3)

rhog_i = fromfile(f'{path_tapwind_jup_e3}gasdens0.dat').reshape(ny,nx)    
rhod_i = fromfile(f'{path_tapwind_jup_e3}dust1dens0.dat').reshape(ny,nx)
```

```python
def vortensity_e3456_movie(param):
    
    fpath, paths, nf = param
    
    diff_rad = np.diff(rad_dat)         # do this for log scale w Jaehan
    diff_phi = np.diff(phi_dat)         # here, do this instead dr = rad_dat[1] - rad_dat[0]
                                        # dphi = phi_dat[1] - phi_dat[0]
    kind = ['tapwind_sat_e3','tapwind_sat_e4','tapwind_sat_e5','tapwind_sat_e6']
    rhog_arr  = []
    gas_rhog_arr = []
    vortensity_arr = []
    gas_vortensity_arr = []

    for name in kind:
        
        rhog_i = fromfile(f'{fpath}{name}/gasdens0.dat').reshape(ny,nx)
        rhog   = fromfile(f'{fpath}{name}/gasdens{nf}.dat').reshape(ny,nx)
        gas = rhog/rhog_i
        gas_rhog = gas*rhog

        omega = np.sqrt(1/rad[:,None]**3)
        omega0 = 1

        vr  = fromfile(f'{fpath}{name}/gasvy{nf}.dat').reshape(ny,nx)
        vr0 = fromfile(f'{fpath}{name}/gasvy0.dat').reshape(ny,nx)

        vphi  = fromfile(f'{fpath}{name}/gasvx{nf}.dat').reshape(ny,nx)
        vphi0 = fromfile(f'{fpath}{name}/gasvx0.dat').reshape(ny,nx)
        vphik = np.power(rad, -0.5)
        vframe = rad * (np.power(rad, -1.5) - omega0)
        vphi += (vphik - vframe)[:,None]

        sound_speed = 0.06*rad[:,None]*omega   # c_s
        append = diff_rad[-1]*vphi[-1,:]
        prepend = np.transpose(vr[:,-1]-vr0[:,-1])[:,None]
        first_term  = np.diff(rad[:,None]*vphi[:,:], axis=0, append=append[None,:]) 
        first_term /= (rad*diff_rad)[:,None]
        second_term = np.diff(vr[:,:] - vr0[:,:], axis=-1, prepend=prepend) / phi[None,:]

        del_v = (first_term - second_term) / rad[:,None]
        vorticity = del_v   # vorticity
        vortensity = (np.power(sound_speed, 2) * (rhog)/vorticity)/rad[:,None]  # vortensity
        gas_vortensity = gas*vortensity

        rhog_arr.append(rhog)
        gas_rhog_arr.append(gas_rhog)
        vortensity_arr.append(vortensity)
        gas_vortensity_arr.append(gas_vortensity)

    fig, ax1 = plt.subplots(nrows=1, ncols=1 ,figsize=(8,6))
    ax2 = ax1.twinx()
    
    ax1.plot(rad, np.median(vortensity_arr[0], axis=1), label=f'{kind[0]}', c='darkviolet', linewidth=3)
    plt.axvline(x=rad[np.argmax(np.mean(gas_vortensity_arr[0], axis=1))], label=f'{kind[0]}', c='darkviolet', ls='--')
    ax1.plot(rad, np.median(vortensity_arr[1], axis=1), label=f'{kind[1]}', c='violet', linewidth=3)
    plt.axvline(x=rad[np.argmax(np.mean(gas_vortensity_arr[1], axis=1))], label=f'{kind[1]}', c='violet', ls='--')
    
    ax1.plot(rad, np.median(vortensity_arr[2], axis=1), label=f'{kind[2]}', c='red', linewidth=3)
    plt.axvline(x=rad[np.argmax(np.mean(gas_vortensity_arr[2], axis=1))], label=f'{kind[2]}', c='red', ls='--')
    ax1.plot(rad, np.median(vortensity_arr[3], axis=1), label=f'{kind[3]}', c='orange', linewidth=3)
    plt.axvline(x=rad[np.argmax(np.mean(gas_vortensity_arr[3], axis=1))], label=f'{kind[3]}', c='orange', ls='--')
    
    ax2.set_title(fr't={nf}, h/r=0.06, $\alpha\approx 3\times 10^{-5}$, $m_p=3\times 10^{-4}M_j$', fontsize=20)
    ax2.set_ylabel(r'$L_{iso} \equiv c^{2}_{s} \Sigma/(\bigtriangledown \times v)_{z}$', fontsize=20)
    ax2.tick_params(axis="y", labelcolor='darkviolet', labelsize=18)
    ax2.set_ylim(0, 0.025)
    ax1.set_xlabel(r'Radius [$r_p$]', fontsize=20)
    ax1.tick_params(axis="x", labelsize=18)
    ax1.tick_params(axis="y", labelcolor='violet', labelsize=18)
    ax1.set_ylim(0, 0.025)
    ax1.set_xlim(1,1.5)
    ax1.legend(loc='upper right')
    
    plt.savefig(f'{paths}/vortensity_e3456_long_{nf}.png', transparent=True, dpi=300, bbox_inches='tight') # dots per inch
    plt.close()
```

```python
def vortensity_e3456_noise_canceling_movie(param):
    
    fpath, paths, name, nf = param
    
    diff_rad = np.diff(rad_dat)         # do this for log scale w Jaehan
    diff_phi = np.diff(phi_dat)         # here, do this instead dr = rad_dat[1] - rad_dat[0]
                                        # dphi = phi_dat[1] - phi_dat[0]
    kind = ['tapwind_sat_e3','tapwind_sat_e4','tapwind_sat_e5','tapwind_sat_e6']
    rhog_arr  = []
    gas_rhog_arr = []
    vortensity_arr = []
    gas_vortensity_arr = []
    noise_canceled_arr = []
    noise_canceled_arr2 = []

    for name in kind:
        
        rhog_i = fromfile(f'{fpath}{name}/gasdens0.dat').reshape(ny,nx)
        rhog   = fromfile(f'{fpath}{name}/gasdens{nf}.dat').reshape(ny,nx)
        gas = rhog/rhog_i
        gas_rhog = gas*rhog

        omega = np.sqrt(1/rad[:,None]**3)
        omega0 = 1

        vr  = fromfile(f'{fpath}{name}/gasvy{nf}.dat').reshape(ny,nx)
        vr0 = fromfile(f'{fpath}{name}/gasvy0.dat').reshape(ny,nx)

        vphi  = fromfile(f'{fpath}{name}/gasvx{nf}.dat').reshape(ny,nx)
        vphi0 = fromfile(f'{fpath}{name}/gasvx0.dat').reshape(ny,nx)
        vphik = np.power(rad, -0.5)
        vframe = rad * (np.power(rad, -1.5) - omega0)
        vphi += (vphik - vframe)[:,None]

        sound_speed = 0.06*rad[:,None]*omega   # c_s
        append = diff_rad[-1]*vphi[-1,:]
        prepend = np.transpose(vr[:,-1]-vr0[:,-1])[:,None]
        first_term  = np.diff(rad[:,None]*vphi[:,:], axis=0, append=append[None,:]) 
        first_term /= (rad*diff_rad)[:,None]
        second_term = np.diff(vr[:,:] - vr0[:,:], axis=-1, prepend=prepend) / phi[None,:]

        del_v = (first_term - second_term) / rad[:,None]
        vorticity = del_v   # vorticity
        vortensity = (np.power(sound_speed, 2) * (rhog)/vorticity)/rad[:,None]  # vortensity
        gas_vortensity = gas*vortensity
        
        # noise canceling
        n = 21 # window_length must be odd number
        yy = savgol_filter(np.mean(vortensity, axis=1), n, 2)
        yyy = savgol_filter(np.mean(gas_vortensity, axis=1), n, 2)
        
        rhog_arr.append(rhog)
        gas_rhog_arr.append(gas_rhog)
        vortensity_arr.append(vortensity)
        gas_vortensity_arr.append(gas_vortensity)
        noise_canceled_arr.append(yy)
        noise_canceled_arr2.append(yyy)

    fig, ax1 = plt.subplots(nrows=1, ncols=1 ,figsize=(8,6))
    ax2 = ax1.twinx()

    ax1.plot(rad, noise_canceled_arr[0], label=f'{kind[0]}', linewidth=3, c='darkviolet')
    ax1.plot(rad, noise_canceled_arr[1], label=f'{kind[1]}', linewidth=3, c='violet')
    ax1.plot(rad, noise_canceled_arr[2], label=f'{kind[2]}', linewidth=3, c='red')
    ax1.plot(rad, noise_canceled_arr[3], label=f'{kind[3]}', linewidth=3, c='orange')
    plt.axvline(x=rad[np.argmax(noise_canceled_arr2[0])], label=f'{kind[0]}', c='darkviolet', ls='--')
    plt.axvline(x=rad[np.argmax(noise_canceled_arr2[1])], label=f'{kind[1]}', c='violet', ls='--')
    plt.axvline(x=rad[np.argmax(noise_canceled_arr2[2])], label=f'{kind[2]}', c='red', ls='--')
    plt.axvline(x=rad[np.argmax(noise_canceled_arr2[3])], label=f'{kind[3]}', c='orange', ls='--')

    ax1.set_title(fr't={nf}, h/r=0.06, $\alpha\approx 3\times 10^{-5}$, $m_p=3\times 10^{-4}M_j$', fontsize=18)
    ax1.set_ylabel(r'$L_{iso} \equiv c^{2}_{s} \Sigma/(\bigtriangledown \times v)_{z}$', fontsize=20)
    ax1.tick_params(axis="y", labelcolor='darkviolet', labelsize=18)
    ax1.set_xlabel(r'Radius [$r_p$]', fontsize=20)
    ax1.tick_params(axis="x", labelsize=18)
    ax1.set_ylim(0, 0.025)
    ax1.set_xlim(1,1.5)
    ax1.legend(loc='best')
    
    plt.savefig(f'{paths}/vortensity_e3456_noise_canceling_{nf}.png', transparent=True, dpi=300, bbox_inches='tight') # dots per inch
    plt.close()
```

```python
def rossby_movie(param):
    
    fpath, path, name, nf = param
    
    kind = ['gas','dust1']
    arr  = []

    for k in kind:
        
        diff_rad = np.diff(rad_dat)         # do this for log scale w Jaehan
        diff_phi = np.diff(phi_dat)         # here, do this instead dr = rad_dat[1] - rad_dat[0]
                                            # dphi = phi_dat[1] - phi_dat[0]

        omega = np.sqrt(1/rad[:,None]**3)

        vr  = fromfile(f'{fpath}{name}/{k}vy{nf}.dat').reshape(ny,nx)
        vr0 = fromfile(f'{fpath}{name}/{k}vy0.dat').reshape(ny,nx)
        
        vphi  = fromfile(f'{fpath}{name}/{k}vx{nf}.dat').reshape(ny,nx)
        vphi0 = fromfile(f'{fpath}{name}/{k}vx0.dat').reshape(ny,nx)
        
        append = diff_rad[-1]*vphi[-1,:] - diff_rad[-1]*vphi0[-1,:]
        prepend = np.transpose(vr[:,-1]-vr0[:,-1])[:,None]
        
        first_term  = np.diff(rad[:,None]*vphi[:,:] - rad[:,None]*vphi0[:,:], axis=0, append=append[None,:]) 
        first_term /= (rad*diff_rad)[:,None]
        second_term = np.diff(vr[:,:] - vr0[:,:], axis=-1, prepend=prepend) / phi[None,:]
        
        del_v = (first_term - second_term) / rad[:,None]
        rossby = del_v / 2*omega
        
        arr.append(rossby)

    fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(20,6), sharey=True)
    
    for i, ax in zip(arr, axes.flat):
        rossby_plot = ax.pcolormesh(phi*180/np.pi + 90, rad, i, cmap= "magma")
    
        cmin = -0.15
        cmax = 0.0
        rossby_plot.set_clim([cmin, cmax])
        
        ax.set_ylim(0.8,1.8)
        ax.set_xlim(0,360)
        ax.set_xticks(np.arange(0, 420, step=60))
#         ax.set_aspect('equal')

        cbar = fig.colorbar(rossby_plot, ax=ax)
        fig.tight_layout(pad=0.05)
        
    axes[0].set_title(f'gas: {nf}', fontsize=20)
    axes[1].set_title(f'dust: {nf}', fontsize=20)
    axes[0].set_ylabel('rad [rp]', fontsize=20)
    axes[0].set_xlabel(r'$\phi$ (Degree)', fontsize=20)
    axes[1].set_xlabel(r'$\phi$ (Degree)', fontsize=20)  
#     cbar = fig.colorbar(rossby_plot, ax=axes[1:])
    cbar.set_label(r'$\Sigma / \Sigma_{0}$', fontsize=20)

    plt.savefig(f'{path}{name}/rossby/rossby_{nf}.png', transparent=True, dpi=300, bbox_inches='tight') # dots per inch
    plt.close()
    
name = ['cooling_sat_e4']

# input for parallel jobs for MassTaper torque and wind
for i in name:
    num_cores = 12
    frame_range = range(0, 1399, 2)
    params = [(pathc, paths, i, nf) for nf in frame_range]
    
    # Iterate through frames
    if len(frame_range) == 1:
        rossby_movie(params[0])
    else:
        if num_cores > 1:
            p = Pool(num_cores) # default number of processes is multiprocessing.cpu_count()
            p.map(rossby_movie, params)
            p.terminate()
        else:
            for param in params:
                rossby_movie(param)
                
    base_path = f'{paths}{i}/rossby'
    base_name = 'rossby_'
    
    old_range = frame_range
    new_range = range(0, len(old_range))
    
    for (old_number, new_number) in zip(old_range, new_range):
            old_file = "%s/%s%d.png" % (base_path, base_name, old_number)
            tmp_file = "%s/tmp_%s%d.png" % (base_path, base_name, new_number)
            shutil.copyfile(old_file, tmp_file)
```
